```python
import nltk
nltk.download('book')
nltk.download('machado')
nltk.download('mac_morpho')
from nltk.corpus import machado

livro_raw = ""

for id in machado.fileids():
    livro_raw += (machado.raw(id))
```

```python
import nltk
from urllib import request

urls = ["http://www.gutenberg.org/cache/epub/17591/pg17591.txt","http://www.gutenberg.org/cache/epub/24411/pg24411.txt" ,"http://www.gutenberg.org/cache/epub/24401/pg24401.txt", "http://www.gutenberg.org/cache/epub/44540/pg44540.txt","http://www.gutenberg.org/cache/epub/29040/pg29040.txt" ,"http://www.gutenberg.org/cache/epub/38496/pg38496.txt"]

for url in urls:
    livro = request.urlopen(url)
    livro_raw += livro.read().decode('utf-8-sig') 
    
print(len(livro_raw))
print(type(livro_raw))
```

```python
from nltk import word_tokenize
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
livro_raw = livro_raw.lower()
sentences = nltk.sent_tokenize(livro_raw)
print(sentences[:30])
```

```python
tokens_sentences = []
for i in sentences:
    
    tokens_sentences.append(word_tokenize(i))
print(tokens_sentences[:30])
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
tl2=[]
for sentence in tokens_sentences:
    tl2_sentence=[]
    for word in sentence:
        if word not in stopwords:
            tl2_sentence.append(word)
    tl2.append(tl2_sentence)

saida = []
for t in tl2:
    saida.append([word for word in t if word.isalpha()])
print(saida[:30])
```

```python
livro = []
stemmer = nltk.stem.RSLPStemmer()
for sentence in saida:
    sentences_strem=[]
    for word in sentence:
        sentences_strem.append(stemmer.stem(word))
    livro.append(sentences_strem)
print(livro[:30])
```

```python
from gensim.models import Word2Vec

# sem nenhum outro parametro
#print(type(words))
model = Word2Vec(livro)
#model.save("word2vec.model")
vocabulary = list(model.wv.vocab)
print(vocabulary)
```

```python
similar_words = model.wv.most_similar("crianç")
print(similar_words)
```

```python
model = Word2Vec(livro, min_count = 2, window = 2, size = 50, sample=0, alpha=0.05,  min_alpha=0.0007, negative=5, workers=2)
vocabulary = list(model.wv.vocab)
print(vocabulary)
```

```python
similar_words = model.wv.most_similar("crianç")
print(similar_words)
```
