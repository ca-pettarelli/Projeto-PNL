```python
from nltk.stem import RSLPStemmer
```

```python
def Tokenize(sentence):
    sentence = sentence.lower()
    sentence = nltk.word_tokenize(sentence)
    return sentence
```

```python
def Stemming(sentence):
    stemmer = RSLPStemmer()
    phrase = []
    for word in sentence:
        phrase.append(stemmer.stem(word.lower()))
    return phrase
```

```python
def RemoveStopWords(sentence):
    stopwords = nltk.corpus.stopwords.words('portuguese')
    phrase = []
    for word in sentence:
        if word not in stopwords:
            phrase.append(word)
    return phrase
```

```python
import nltk
nltk.download('book')
nltk.download('machado') 
nltk.download('mac_morpho')
```

```python
from nltk.corpus import machado   #importando todos os livros de Machado

livro_raw = ""

for id in machado.fileids():
    livro_raw += (machado.raw(id))   #colocando todos os livro na string

print(livro_raw[:30])
print(len(livro_raw))

# imports
import nltk
from nltk import word_tokenize
stopwords = nltk.corpus.stopwords.words('portuguese')

#colocando todas as palavras em minusculo
livro_raw = livro_raw.lower()

#tokenizando as sentenças
sentences = nltk.sent_tokenize(livro_raw)

print(sentences[:30])
```

```python
#tokenizando cada palavra de cada sentença

tokens_sentences = []
for i in sentences:
    tokens_sentences.append(word_tokenize(i))
    
print(tokens_sentences[:30])
```

```python
#imports
stopwords = nltk.corpus.stopwords.words('portuguese')

#retirando as stopwords de cada frase
tl2=[]
for sentence in tokens_sentences:
    tl2_sentence=[]
    for word in sentence:
        if word not in stopwords:
            tl2_sentence.append(word)
    tl2.append(tl2_sentence)

book = []
for t in tl2:
    book.append([word for word in t if word.isalpha()])
print(book[:30])
```

```python
#imports
stemmer = nltk.stem.RSLPStemmer()

# aplicando o stemming em cada palavra de cada frase
livro = []
for sentence in book:
    sentences_strem=[]
    for word in sentence:
        sentences_strem.append(stemmer.stem(word))
    livro.append(sentences_strem)
    
    
print(livro[:30])
```

```python
from gensim.models import Word2Vec

# sem nenhum outro parametro
model = Word2Vec(livro)

vocabulary = list(model.wv.vocab)
print(vocabulary)
```
