```python
import nltk
from nltk import word_tokenize

#colocando todas as palavras em minusculo
produtos = produtos.lower()

# #tokenizando as sentenças
sentences = nltk.sent_tokenize(produtos)

print(sentences[:30])
```

```python
sentences = [i.replace('nan', '') for i in sentences] 
```

```python
len(sentences)    
```

```python
#tokenizando cada palavra de cada sentença
tokens_sentences = []
for i in sentences:

    tokens_sentences.append(word_tokenize(i))
    
    
print(tokens_sentences[:30])
```

```python
#imports
stopwords = nltk.corpus.stopwords.words('portuguese')

#retirando as stopwords de cada frase
tl2=[]
for sentence in tokens_sentences:
    tl2_sentence=[]
    for word in sentence:
        if word not in stopwords:
            tl2_sentence.append(word)
    tl2.append(tl2_sentence)

```

```python
book = []
for t in tl2:
    book.append([word for word in t if word.isalpha()])
print(book[:30])
len(book)
```

```python
from gensim.models import Word2Vec

model = Word2Vec(book, min_count = 2, window = 2, size = 50, sample=0, alpha=0.05,  min_alpha=0.0007, negative=5, workers=2, sg=1)
vocabulary = model.wv.vocab
print(len(vocabulary))
similar_words = model.wv.most_similar('notebook')
print(similar_words)
```
